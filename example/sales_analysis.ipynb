# Sales Data Analysis

This notebook analyzes quarterly sales data to identify trends and insights.

## Data Loading

First, we load the sales data and examine its structure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create sample sales data
data = {
    'month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'product': ['Widget A', 'Widget B', 'Widget A', 'Widget B', 'Widget A', 'Widget B'],
    'region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'sales': [150, 200, 175, 225, 200, 250],
    'price': [29.99, 49.99, 29.99, 49.99, 29.99, 49.99]
}
df = pd.DataFrame(data)
df.head()
```

## Data Cleaning

Check for missing values and data types.

```python
print("Missing values:")
print(df.isnull().sum())
print("\nData types:")
print(df.dtypes)
```

```python
# Calculate revenue
df['revenue'] = df['sales'] * df['price']
df.head(3)
```

## Visualization

Let's visualize the sales trends.

```python
# Plot monthly sales by product
plt.figure(figsize=(8, 4))
for product in df['product'].unique():
    product_data = df[df['product'] == product]
    plt.plot(product_data['month'], product_data['sales'], marker='o', label=product)

plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.title('Monthly Sales by Product')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

Key findings from the analysis.

```python
print("=== Sales Summary ===")
print(f"\nTotal Revenue: ${df['revenue'].sum():,.2f}")
print(f"Total Units Sold: {df['sales'].sum():,}")
print(f"Average Price: ${df['price'].mean():.2f}")
print(f"\nRevenue by Product:")
print(df.groupby('product')['revenue'].sum())
print(f"\nTop performing month: {df.groupby('month')['sales'].sum().idxmax()} with {df.groupby('month')['sales'].sum().max()} units")
```
